```python
import matplotlib.pyplot as plt
from sklearn import datasets
import numpy as mp
import pandas as pd
import matplotlib.pyplot as plt


```

```python
iris = datasets.load_iris()
```

```python
irisInputArray = iris.data
irisOutputArray = iris.target
```

```python
irisInputDf = pd.DataFrame(irisInputArray)
irisInputDf['Output'] = pd.Series(irisOutputArray)
```

```python
irisInputDf
```

```python
plt.plot(irisInputDf[0],irisInputDf[1],'red')

plt.plot(irisInputDf[2],irisInputDf[1], 'blue')
plt.xlabel("X label")
plt.title('TITLE')
plt.show()

```

```python
# USing Axes
# axes ([x_coordinate, y_coordinate, width, height])
# Unit between 0 - 1
plt.axes([0.05,0.05,0.425,0.9])
plt.plot(irisInputDf[0],irisInputDf[1],'red')
plt.xlabel("X label")
plt.title('TITLE')

plt.axes([0.525,0.05,0.425,0.9])

plt.plot(irisInputDf[2],irisInputDf[1], 'blue')
plt.xlabel("X label")
plt.title('TITLE')
plt.show()

```

```python
# Using Subplot (no of rows , no of columns, position
# Using Tight layout - if not mentioned then the graph will be overlapping
# It adds padding

plt.subplot(2,1,1)
plt.plot(irisInputDf[0],irisInputDf[1],'red')
plt.xlabel("X label")
plt.title('TITLE')


plt.subplot(2,1,2)
plt.plot(irisInputDf[2],irisInputDf[1], 'blue')
plt.xlabel("X label")
plt.title('TITLE')

plt.tight_layout()
plt.show()



```

```python
# Customizing Axes
# axis([x_min, x_max, y_min, y_max])
# Individual Axis
#xlim([x_min,x_max])
#Zoom In

plt.plot(irisInputDf[0],irisInputDf[1],'red')
plt.xlabel("X label")
plt.title('TITLE')
plt.axis([5.5,7.0, 2,4])
plt.show()
```
